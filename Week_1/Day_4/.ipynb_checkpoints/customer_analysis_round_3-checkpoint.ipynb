### 7. BONUS: Put all the previously mentioned data transformations into a function.

```python
def data_cleaning(name, file):
    import numpy as np
    import pandas as pd
    import datetime
    
    # read the file
    name = pd.read_csv(file)
    
    # shape
    print("The shape of the file 1: \n", name.shape, "\n")
    
    # standardize the columns
    new_columns = []
    for i in range(len(name.columns)):
        new_columns.append(name.columns[i].lower().replace(" ","_"))
    
    name.columns = new_columns
    
    print("The new columns of the file 1 are: \n", name.columns, "\n")
    
    # columns - numerical
    name_num = name._get_numeric_data()
    ("\n")
    
    # columns categorical
    name_cat = name.drop(columns=name_num.columns)
    
    # check the Nan-values
    name.info()
    ("\n")
    nulls_name = pd.DataFrame(round(name.isna().sum()/len(name),4)*100)
    
    #Datetime format
    
    date_name = pd.DataFrame(name, columns = new_columns)
    date_name['effective_to_date'] = pd.to_datetime(date_name['effective_to_date'], errors='coerce')
    date_name["months"] = pd.DatetimeIndex(date_name["effective_to_date"]).month
    date_name.query('months>=1 & months<=3')
    
    print("\n The numerical columns: \n", name_num, "\n")
    print("\n The categrical columns: \n", name_cat, "\n")
    print("\n The overview about null-values in the rows: \n", nulls_name, "\n")
    print("The head of the date_file as a query for the first qarter: ")
    print("The new datatypes are: ", date_name.dtypes, "\n")
    print(date_name.query('months>=1 & months<=3'))
```

```python
data_cleaning(file1, 'marketing_customer_analysis.csv')
```
